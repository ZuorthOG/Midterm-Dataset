## Dataset Description

### College Basketball Dataset

This dataset contains statistics from NCAA Division I college basketball teams across multiple seasons. Overall, the dataset is used to analyze team performance, statistics, and trends in Division I college basketball over multiple seasons.

### Key Features
- **Season Information**: Identifies the year/season each team’s data comes from (2013–2019, 2021–2024, and 2025 updates).

- **Team Name**: The Division I college basketball team the statistics belong to.

- **Conference**: The conference each team competes in

- **Wins and Losses**: Total number of games won and lost during the season.

- **Offensive Statistics** : Metrics that measure how effective a team is at scoring, such as offensive efficiency and shooting percentages.

- **Defensive Statistics** : Measures of how well a team prevents opponents from scoring, including defensive efficiency.

- **Advanced Metrics**: Analytics such as effective field goal percentage, turnover rates, rebound rates, and strength of schedule.

- **Postseason Indicators**: Some seasons include information about tournament appearances or rankings.

### Column Descriptions
- **`RK`**: Name of the Division I college basketball team.

- **`CONF`**: The athletic conference the team belongs to.

- **`G`**: Total number of games played in the season.

- **`W`**: Total number of games won.

- **`ADJOE`**: Adjusted Offensive Efficiency

- **`ADJDE`**: Adjusted Defensive Efficiency

- **`BARTHAG`**: Power rating representing the probability of beating an average Division I team.

- **`EFG_O`**: Effective field goal percentage for the team’s offense.

- **`EFG_D`**: Effective field goal percentage allowed by the team’s defense.

- **`TOR`**: Turnover percentage allowed (turnover rate).

- **`TORD`**: Turnover percentage committed (steal rate by opponents).

- **`ORB`**: Offensive rebound rate.

- **`DRB`**: Offensive rebound rate allowed (defensive rebounding ability of opponents).

- **`FTR`**: Free throw rate (how often the team gets to the free throw line).

- **`FTRD`**: Free throw rate allowed.

- **`2P_O`**: Two-point shooting percentage.

- **`2P_D`**: Two-point shooting percentage allowed.

- **`3P_O`**: Three-point shooting percentage.

- **`3P_D`**: Three-point shooting percentage allowed.

- **`ADJ_T`**: Adjusted tempo (estimated possessions per 40 minutes against an average tempo team).

- **`WAB`**: Wins Above Bubble (how a team performs relative to the NCAA tournament cutoff).

- **`POSTSEASON`**: The round where the team was eliminated in the NCAA tournament (e.g., R64, S16, F4, Champion).

- **`SEED`**: Tournament seed assigned to the team.

- **`YEAR`**: The season year.

**Source**: [Kaggle Datasets](https://www.kaggle.com/datasets/andrewsundberg/college-basketball-dataset)
